# Transforms
データはいつも学習データに求められる型になっているとは限らない．transformsを使い学習に適したデータへ変換する．

全てのTorchVisionデータセットには２つのパラメータがある．`transform`(特徴量を修正する)と`target_transform`(ラベルを修正する)がある．`torchvision.transforms`

FashionMNISTは．PILイメージフォーマットでラベルは整数である．学習時には特徴量を正規化テンソルとして使い，ラベルはone-hotエンコーディングする必要がある．この変換には，`ToTensor`および`Lambda`を使用する．

````python
import torch 
from torchvision import datasets 
from torchvision.transforms import ToTensor, Lambda

ds = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True, 
    transform=ToTensor(),
    target_transform=Lambda(lambda y:  torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)
````

### ToTensor()
`ToTensor`は，`ndarray`などを`FloatTensor`に変換する．また，ピクセル値を[0., 1.]に収める．


### Lambda Transforms
ここでは，無名関数を使ってone-hotエンコードされたテンソルを作成している．サイズ10のテンソルを作り，`scatter_`を使用してインデックスに応じて`value=1`が入ったラベル`y`を作る．

```
target_transform = Lambda(lambda y: torch.zeros(
    10, dtype=torch.float).scatter_(dim=0, index=torch.tensor(y), value=1))
```
